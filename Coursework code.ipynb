```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

### Joining the country names data by country code

```python
country = pd.read_excel('Country-Code.xlsx')
data = pd.merge(zomato, country, on='Country Code')
data.head()
```

### Grouping the country code to get the number of restaurants

```python
country_data = data.groupby(['Country'], as_index=False).count()[['Country', 'Restaurant ID']]
country_data.head()
country_data.columns = ['Country', 'No of Restaurant']
```

### Plotting the number of restaurants by country

```python
plt.bar(country_data['Country'], country_data['No of Restaurant'])
plt.xlabel('Countries', fontsize=25)
plt.rcParams['figure.figsize']=(30,20) 
plt.ylabel('No. of Restaurants', fontsize=25)
plt.title('Number of restaurants by Country', fontsize=30)
```

### Plotting the number of restaurants by cities of India

```python
city_data = data[data['Country'] =='India']
city_data1 = city_data['City'].value_counts()
city_data1.plot.bar(figsize=(14, 7),fontsize=14)
plt.title('No. of restaurants by city (India)', fontsize=30)                                             
plt.xlabel('City', fontsize=20)
plt.ylabel('No. of Restaurants', fontsize=20)
```

### Mapping the spacial distribution of restaurants on Zomato

```python
xmm = (data.Longitude.min()+data.Longitude.max())/2
ymm = (data.Latitude.min()+data.Latitude.max())/2
```

```python
import folium
from folium.plugins import HeatMap
```

```python
map = folium.Map(location=[ymm,xmm], zoom_start=6, width=600,height=700)
data.apply(lambda row:folium.CircleMarker(location=[row["Latitude"], row["Longitude"]], 
                                        radius=3, opacity=.05, color='#cc0000').add_to(map), axis=1)
map
```

### Replace the dots by a density map (a.k.a. heatmap)

This operation represents the spatial distribution as a smooth continuous surface.

```python
hmap = folium.Map(location=[ymm,xmm], zoom_start=6, width=600,height=700)
hm_wide = HeatMap(list(zip(data.Latitude.values, data.Longitude.values)), 
                  min_opacity=0.2,
                  radius=10, blur=15, 
                  max_zoom=1)
hmap.add_child(hm_wide)
```

### Change the colour scale and the level of smoothing

We replace the "rainbow" colour scale, which is not recommended for representing numeric data, by a better one.
The smoothing level is controlled by the value of the parameter 'radius'. Change the value of this parameter to observe its impact on the visible patterns.

```python
hmap = folium.Map(location=[ymm,xmm], zoom_start=6, width=600,height=700)
hm_wide = HeatMap(list(zip(data.Latitude.values, data.Longitude.values)), 
                  min_opacity=0.2,
                  #radius=5,
                  radius=4.5,
                  blur=10, 
                  #Adjusting the radius will make areas more distict, and reducing blur will make map clearer!
                  max_zoom=1, gradient={.5: 'blue', .75: 'orange', 0.9: 'red'})
hmap.add_child(hm_wide)
```

### Building a linear regression model

```python
in_data = india_data.drop(columns=['Restaurant ID','Restaurant Name','Country Code', 'City', 'Address', 'Locality', 'Locality Verbose','Longitude', 'Latitude', 'Cuisines', 'Currency', 'Has Table booking', 'Has Online delivery', 'Is delivering now','Switch to order menu','Rating color', 'Rating text', 'Country'])
```

```python
in_data.head()
```

```python
X = in_data.drop(columns=['Aggregate rating']).values
Y = in_data['Aggregate rating'].values
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,random_state=0)
```

```python
print(X_train.shape)
print(X_test.shape)
print(Y_train.shape)
print(Y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression

reg = LinearRegression()

reg.fit(X_train,Y_train)
```

```python
y_predict = reg.predict(X_test)
```

```python
from sklearn.metrics import r2_score
r2_score(Y_test,y_predict)
```

```python
import statsmodels.api as sm
from statsmodels.formula.api import ols
```

```python
residuals = sns.residplot(x=y_predict, y=Y_test, lowess=True, color="g")
```

### Improving the accuracy of linear regression model

```python
print("lr.coef_: {}".format(reg.coef_))
print("lr.intercept_: {}".format(reg.intercept_))
print("Training set score: {:.2f}".format(reg.score(X_train, Y_train)))
print("Test set score: {:.7f}".format(reg.score(X_test, Y_test)))
```

### Fitting the data to a random forest regressor model to see if it improves accuracy

```python
from sklearn.linear_model import RidgeCV, LinearRegression, Lasso
from sklearn.ensemble import RandomForestRegressor

reg = RandomForestRegressor(n_estimators=100)
```

```python
reg.fit(X_train, Y_train)
```

```python
print("Training set score: {:.2f}".format(reg.score(X_train, Y_train)))
print("Test set score: {:.2f}".format(reg.score(X_test, Y_test)))
```
